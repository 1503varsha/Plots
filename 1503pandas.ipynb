```python
#Drops problematic rows
df=df.dropna()
```

```python
rooms=df["Rooms"]
print(rooms)
```

```python
price=df["Price"]
print(price)
```

```python
import matplotlib.pyplot as plt
```

```python
print(min(price))
print(max(price))
print((max(price)-min(price))/10)
```

```python
fig=plt.figure()
plt.hist(x=price, bins=90, edgecolor="orange", color="black")
plt.title('Histogram')
plt.xlabel('Price range')
plt.ylabel('no. of houses')
plt.show()
```

```python
rooms_count=df["Rooms"].value_counts()
```

```python
rooms_count=dict(rooms_count)
```

```python
rooms_count
```

```python
roomsize=list(rooms_count.keys())
no_rooms=list(rooms_count.values())
```

```python
fig=plt.figure(figsize=(8,8))
plt.pie(x=no_rooms, labels=roomsize,)
plt. legend()
plt.title('PieChart')
plt.show()
```

```python
fig=plt.figure(figsize=(8,8))
plt.plot(no_rooms,roomsize)

plt.title('PieChart')
plt.show()
```

```python
rn=df["Regionname"].value_counts()
rn=dict(rn)
print(rn)
k=list(rn.keys())
v=list(rn.values())
```

```python
fig=plt.figure(figsize=(7,7))
plt.pie(x=v, labels=k)
plt.legend()
plt.title('PieChart')
plt.show()
```

```python
typ=df["Type"].value_counts()
typ=dict(typ)
print(typ)
k=list(typ.keys())
v=list(typ.values())
plt.bar(k,v)
plt.grid(True)
plt.show() 

```

```python
mc=df["Method"].value_counts()
mc=dict(mc)
print(mc)
k=mc.keys()
v=mc.values()
plt.errorbar(k,v, fmt = '.-')
plt.show()
```

```python
su=df["Suburb"].value_counts()
su=dict(su)

k=su.keys()
v=su.values()
fig=plt.figure(figsize=(17,17))
plt.bar(x=v,height=k)
plt.show()
```

```python
mc=df["Method"].value_counts()
mc=dict(mc)
print(mc)
k=mc.keys()
v=mc.values()
plt.fill_between(k,v)
plt.show()
```
